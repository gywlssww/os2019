# TD-IDF 

```python
import pandas as pd

cleaned_df=pd.read_csv('../data/mixed_all_data.csv', delimiter='\t')
doc_list=[eval(doc_str) for doc_str in cleaned_df['doc_txt'].tolist()]
tdidf_doc_list=[" ".join(doc) for doc in doc_list]
```

Parameter

```python
unigram_range = (1,1)
bigram_range=(2,2)
min_df = 10
max_df = 1.
max_features = 4286
```

## Uni-gram

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_uni = TfidfVectorizer(encoding='utf-8',
                        ngram_range=unigram_range,
                        stop_words=None,
                        lowercase=False,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2',
                        sublinear_tf=True)
```

- uni_features : scipy.sparse.csr.csr_matrix 
- uni_word2idx : word_to_idx dict
- uni_idx2word : idx_to_word dict

```python
uni_features=tfidf_uni.fit_transform(tdidf_doc_list) #test set 은 transform만
uni_word2idx=tfidf_uni.vocabulary_
uni_idx2word={word: ii for ii, word in tfidf_uni.vocabulary_.items()}
```

```python
uni_df= pd.DataFrame(uni_features.toarray(), columns=tfidf_uni.get_feature_names())
uni_df['dep']=cleaned_df['dep'].tolist()
uni_df
```

```python
idx=[0]
uni_word2idx_df=pd.DataFrame(uni_word2idx,index=idx)
uni_word2idx_df
```

## Bi-gram

```python
tfidf_bi = TfidfVectorizer(encoding='utf-8',
                        ngram_range=bigram_range,
                        stop_words=None,
                        lowercase=False,
                        max_df=max_df,
                        min_df=min_df,
                        max_features=max_features,
                        norm='l2',
                        sublinear_tf=True)
```

- bi_features : scipy.sparse.csr.csr_matrix 
- bi_word2idx : word_to_idx dict
- bi_idx2word : idx_to_word dict

```python
bi_features=tfidf_bi.fit_transform(tdidf_doc_list) #test set 은 fit만
bi_word2idx=tfidf_bi.vocabulary_
bi_idx2word={word: ii for ii, word in tfidf_bi.vocabulary_.items()}
```

```python
bi_df= pd.DataFrame(bi_features.toarray(), columns=tfidf_bi.get_feature_names())
bi_df['dep']=cleaned_df['dep'].tolist()
bi_df
```
